```python
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')
#visualization with altair? needs these three lines!!! 

#Load the dataset
from vega_datasets import data
cars = data.cars()
```

### Determine Column Headers

```python
print(cars.columns)
```

## Scatterplot 

```python
#create a new vis, give it your new dataframe, specify mark type- the thing you map data to
#mark_point is basically a scatterplot 
#encode is saying we want to specify what attributes we want to represent our data
alt.Chart(cars).mark_point().encode(
    x = 'Horsepower',
    y = 'Miles_per_Gallon',
    color = 'Origin'
).interactive()
```

## Bar Charts
Good for one numerical (Y-axis) and one categorical (X-axis) 

```python
alt.Chart(cars).mark_bar().encode(
    x = 'Name',
    y='Cylinders',
    color= 'Origin'
)
```

### Aggregate things together (using the average)

```python
alt.Chart(cars).mark_bar().encode(
    x='Year',
    y='average(Cylinders)',
    color = 'Origin'
)
```

## Histogram
### Binning 

```python
alt.Chart(cars).mark_bar().encode(
    #define a custom x variable that'll map to my (horsepower) variable and bin it  
    alt.X('Horsepower', bin=True),
    y = 'count()',
    color = 'Origin'
)
```

## More than 2 variables

```python
alt.Chart(cars).mark_point().encode(
    x ='Horsepower',
    y='Acceleration',
    color = 'Origin',
    size = 'Year'
).interactive()
```
