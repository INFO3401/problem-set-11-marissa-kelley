## Problem 7

Generate three additional visualizations using this data (please use different visualization techniques for each visualization). What do these visualizations tell you about potential collaborations between Shakespeare and Marlowe? 

### Visualization 1

```python
alt.Chart(authors, width = 200).mark_bar().encode(
    x = "Author",
    y = "average(Positive)",
    color = "Author"
).interactive()
```

This bar graph shows that Shakespeare's texts have more positive sentiment, despite being known for his tragedies. 

## Visualization 2

```python
alt.Chart(authors).mark_point().encode(
    x= "average(Neutral)",
    y = "Text",
    color = "Author"
).interactive()
```

This visualzation shows that both authors have a pretty standard amount of neutral writing within all of their texts. 

### Visualization 3

```python
alt.Chart(authors, width = 200).mark_line().encode(
    x = "average(Negative)",
    y = "Text",
    color = "Author")
```

This line graph shows the average amount of negative sentiment within the texts. The text authors are colored differently as to show who wrote what. It's not the most useful graph, but does show that Shakespeare seemed to write both incredibly negative and less negative texts, whereas Marlowe stayed pretty average with his texts.
